```python
import matplotlib.pyplot as plt
import numpy as np

def hitung_posisi(v0, a, t):
    return v0 * t + 0.5 * a * t * t

def hitung_kecepatan_akhir(v0, a, t):
    return v0 + a * t

def tampilkan_menu():
    print("========== Program Gerak Lurus Beraturan (GLBB) ==========")
    print("1. Hitung Posisi Akhir")
    print("2. Hitung Kecepatan Akhir")
    print("3. Keluar")
    print("=======================================================")

def main():
    file_output = open("output.txt", "w")

    if file_output.closed:
        print("Error: Tidak dapat membuka file output.")
        return 1

    pilihan = 0
    v0, a, t = 0, 0, 0

    waktu = []
    posisi_akhir = []
    kecepatan_akhir = []

    while pilihan != 3:
        tampilkan_menu()

        try:
            pilihan = int(input("Pilih menu (1-3): "))
        except ValueError:
            print("Input tidak valid. Masukkan angka 1-3.")
            continue

        if pilihan == 1:
            v0 = float(input("Masukkan kecepatan awal (v0): "))
            a = float(input("Masukkan percepatan (a): "))
            t = float(input("Masukkan waktu (t): "))

            posisi = hitung_posisi(v0, a, t)
            print(f"Posisi akhir: {posisi:.2f}")

            file_output.write(f"Posisi akhir: {posisi:.2f}\n")

            waktu.append(t)
            posisi_akhir.append(posisi)
            kecepatan_akhir.append(hitung_kecepatan_akhir(v0, a, t))

        elif pilihan == 2:
            v0 = float(input("Masukkan kecepatan awal (v0): "))
            a = float(input("Masukkan percepatan (a): "))
            t = float(input("Masukkan waktu (t): "))

            kecepatan_akhir = hitung_kecepatan_akhir(v0, a, t)
            print(f"Kecepatan akhir: {kecepatan_akhir:.2f}")

            file_output.write(f"Kecepatan akhir: {kecepatan_akhir:.2f}\n")

            waktu.append(t)
            posisi_akhir.append(hitung_posisi(v0, a, t))
            kecepatan_akhir.append(kecepatan_akhir)

        elif pilihan == 3:
            print("Program selesai.")
        else:
            print("Pilihan tidak valid. Silakan pilih menu 1-3.")

    file_output.close()

    # Visualisasi data menggunakan matplotlib
    plt.figure(figsize=(10, 6))
    plt.subplot(2, 1, 1)
    plt.plot(waktu, posisi_akhir, marker='o')
    plt.title('Grafik Posisi Akhir terhadap Waktu')
    plt.xlabel('Waktu (s)')
    plt.ylabel('Posisi Akhir')

    plt.subplot(2, 1, 2)
    plt.plot(waktu, kecepatan_akhir, marker='o', color='orange')
    plt.title('Grafik Kecepatan Akhir terhadap Waktu')
    plt.xlabel('Waktu (s)')
    plt.ylabel('Kecepatan Akhir')

    plt.tight_layout()
    plt.show()

if _name_ == "_main_":
    main()
```
